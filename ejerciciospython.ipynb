```python
x = 2*np.pi*np.linspace(0,1,100)
y1 = np.cos(x)
y2 = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y1, 'r', label = r'$\cos(\theta)$')    
ax.plot(x,y2, 'b', label = r'$\sin(\theta)$')
ax.set_xlabel(r'$\theta$', fontsize = 16)
ax.set_ylabel(r'$f(\theta)$', fontsize = 20)
ax.set_title(r'\textbf{Two functions in the same figure}')
ax.legend()
plt.show()
```

```python
plt.rcParams['text.usetex'] = True

x = 2*np.pi*np.linspace(0,1,100)
y1 = np.cos(x)
y2 = np.sin(x)
fig, ax = plt.subplots()
ax.plot(x,y1, 'r', label = r'$\cos(\theta)$')    
ax.plot(x,y2, 'b', label = r'$\sin(\theta)$')

intersection1_x = 5*np.pi / 4
intersection1_y = -np.sqrt(2) / 2

intersection2_x = np.pi
intersection2_y = -1

ax.scatter(intersection1_x, intersection1_y, color='blue', s=50, edgecolor='black', zorder=5) #zorder es para que el punto se dibuje sobre las lneas de las funciones coseno y seno , que tienen un zorder predeterminado ms bajo 
ax.scatter(intersection2_x, intersection2_y, color='blue', s=50, edgecolor='black', zorder=5)
ax.annotate(r'minimum', xy=(np.pi, -1), xytext=(np.pi-0.5, -0.6),arrowprops=dict(facecolor='black', shrink=0.005, width=2, headwidth=5), fontsize=12, color='black', fontweight='bold')
ax.text(5*np.pi/4, -np.sqrt(2)/2 +0.1, r'intersection', fontsize=12, color='black', fontweight='bold')
ax.set_xlabel(r'$\theta$', fontsize = 16)
ax.set_ylabel(r'$f(\theta)$', fontsize = 20)
ax.set_xticks([0, 0.5*np.pi, 0.25*np.pi, np.pi, 1.25*np.pi, 1.5*np.pi, 2*np.pi], labels = ["$0$", r"$\frac{\pi}{2}$", r"$\frac{\pi}{4}$", 
                       r"$\pi$", r"$\frac{5\pi}{4}$", r"$\frac{3\pi}{2}$", r"$2\pi$"])
ax.set_title(r'\textbf{Two functions in the same plot}')
ax.legend()
plt.savefig(os.path.join(folder,'Seno_coseno_puntos_marcados.pdf'))  # Saves the graphic in .pdf
plt.show()  
```

```python
plt.rcParams['text.usetex'] = True

x = np.linspace(-6,6,100)
y1 = np.sinc(x/np.pi)
y2 = np.sinc(x)
fig, ax = plt.subplots()
ax.plot(x,y1, 'b', label = r'sinc(x)')
ax.set_xlabel(r'$x$', fontsize = 16)
ax.set_ylabel(r'$sinc(x)$', fontsize = 20)
#Se usa la funcin min de numpy para encontrar el valor mnimo de la funcin y la funcin 
#argmin para encontrar el ndice de ese valor mnimo.Luego usamos el ndice para encontrar la coordenada  de los arrays x e y correspondiente al mnimo.

min1_index = np.argmin(y1)
min1_x = x[min1_index]
min1_y = y1[min1_index]
ax.annotate(r'minimum', xy=(min1_x, min1_y), xytext=(min1_x+2, min1_y+0.05),arrowprops=dict(facecolor='black', shrink=0.005, width=2, headwidth=5), fontsize=12, color='black', fontweight='bold')

ax.legend()


axin1 = ax.inset_axes([0.65, 0.65, 0.30, 0.30])
axin1.plot(x,y2, 'r', label = r'sinc(\pi x)')
axin1.set_xlabel(r'$x/\pi $', fontsize = 6)
axin1.set_ylabel(r'$sinc(x/\pi)$', fontsize = 10)
min2_index = np.argmin(y2)
min2_x = x[min2_index]
min2_y = y2[min2_index]
axin1.annotate(r'minimum', xy=(min2_x, min2_y), xytext=(min2_x+2, min2_y+0.05),arrowprops=dict(facecolor='black', shrink=0.0005, width=1, headwidth=4), fontsize=9, color='black', fontweight='bold')




plt.show()

```
